```python
import numpy as np
import pandas as pd
import torch
from sklearn import manifold
import matplotlib.pyplot as plt
import seaborn as sb

path = "dataframe_v0_mc.dat"
path = "mc_tree_df.dat"

dat = pd.read_pickle(path)

dat.head()
```

```python
plt.figure(figsize=(10,6))
y,x = np.histogram(dat["v0_ks_mass"],100)
yerr = np.sqrt(y)
plt.errorbar(x,y,yerr=yerr,fmt = '.',capsize = 2,label = "label")

plt.show()
```

```python
tsne = manifold.TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tnse_results = tsne.fit_transform(dat)
```

```python
df_subset = {}
df_subset['tsne-2d-one'] = tnse_results[:,0]
df_subset['tsne-2d-two'] = tnse_results[:,1]
plt.figure(figsize=(16,10))
sb.scatterplot(
    x="tsne-2d-one", y="tsne-2d-two",
    palette=sb.color_palette("hls", 10),
    data=df_subset,
    legend="full",
    alpha=0.3
)
```
